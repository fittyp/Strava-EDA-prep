Hmm as suspected my old watch did not record average speed. We could fill this value with a simple calculation between the distance ran and the time ((Distance*1000)/(Moving Time*60) which would give us speed in m/s which is how it is recorded for the other columns. Lets go for it.

```python
#checking number of NANs in average speed atm
df['Average Speed'].isna().sum()
```

There are 456 NAN values to be replaced.

```python
#create a variable for our average speed calculation
avg_speed_calc = (df['Distance']*1000)/(df['Moving Time']*60)

#using .loc to fillna values with our speed calculation
df.loc[:, "Average Speed"] = df.loc[:, "Average Speed"].fillna(avg_speed_calc)
```

```python
#checking number of NAN's left over
df['Average Speed'].isna().sum()
```

NONE! Nice!
